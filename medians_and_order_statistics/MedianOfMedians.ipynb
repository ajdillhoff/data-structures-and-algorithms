```python
def partition_around(A, p, r, x):
    i = p - 1
    for j in range(p, r):
        if A[j] == x:
            A[j], A[r] = A[r], A[j]
        if A[j] < x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[r] = A[r], A[i + 1]
    return i + 1


def select(A, p, r, i, depth=0):
    print(f"Depth: {depth}")
    print("Original: ", A[p:r + 1])
    while (r - p + 1) % 5 != 0:
        for j in range(p + 1, r + 1):
            if A[p] > A[j]:
                A[p], A[j] = A[j], A[p]
        if i == 0:
            return A[p]
        p += 1
        i -= 1
    
    num_groups = (r - p + 1) // 5
    print(f"p = {p}, r = {r}, i = {i}")

    # Sort each group
    for j in range(num_groups):
        group_idxs = [p + j + k * num_groups for k in range(5)]
        for k in range(1, 5):
            key = A[group_idxs[k]]
            l = k - 1
            while l >= 0 and A[group_idxs[l]] > key:
                A[group_idxs[l + 1]] = A[group_idxs[l]]
                l -= 1
            A[group_idxs[l + 1]] = key

        # Print sorted group
        print("[", end="")
        for k in range(4):
            print(A[group_idxs[k]], end=" ")
        print(f"{A[group_idxs[4]]}]")

    # Find the median of medians
    x = select(A, p + 2 * num_groups, p + 3 * num_groups - 1, num_groups // 2, depth + 1)
    print(f"x = {x}")
    q = partition_around(A, p, r, x)
    print(f"q = {q}")
    k = q - p
    print(f"k = {k}")
    print("After partition: ", A[p:r + 1])
    if i < k:
        return select(A, p, q - 1, i, depth + 1)
    elif i > k:
        return select(A, q + 1, r, i - k - 1, depth + 1)
    else:
        return A[q]


# Create an array of 50 elements to test on
import random

random.seed(6)

A = random.sample(range(1, 100), 15)

median = select(A, 0, 14, 7)
print(median)

A.sort()

print(A)

# Verify with numpy
import numpy as np

print(np.median(A))
```
